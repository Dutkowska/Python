```python
import pygame, sys, os
from pygame.locals import *

def windows():
    size=(1000, 500)
    width=size[0]
    height=size[1]
    wind=pygame.display.set_mode((width, height))
    pygame.display.set_caption("Ping Pong")
    return(wind, width, height)
    
def padel():
    pad_width=100
    pad_height=10
    pad=pygame.Surface([pad_width, pad_height])
    pad_col=(0,98,98)
    pad.fill(pad_col)
    
    pad_rec=pad.get_rect()
    pad_pos=(450, 450)
    pad_rec.x=pad_pos[0]
    pad_rec.y=pad_pos[1]
    return(pad, pad_width, pad_rec, pad_rec.x)

"""def legend():
    leg_width=80
    leg_height=50
    leg=pygame.Surface([leg_width, leg_height])
    leg_col=(0,0,0)
    leg.fill(leg_col)
    leg_rec=leg.get_rect()
    leg_rec.x=900
    leg_rec.y=10
    return(leg, leg_rec)"""

def events(b_speed_x, b_speed_y):
    for event in pygame.event.get():
        if event.type==QUIT:
            pygame.quit()
            sys.exit()
            
        if event.type == pygame.KEYDOWN and event.key == pygame.K_ESCAPE:
            pygame.quit()
            sys.exit()
            
        if event.type==MOUSEMOTION:
            mx, my=event.pos
            pos=mx-(pad_width /2)

            if pos>width - pad_width: #right
                pos=width-pad_width
            if pos<0: #left
                pos=0

            pad_rec.x=pos
            
        if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            b_speed_y=0
            b_speed_x=0
            
        if event.type == pygame.KEYDOWN and event.key == pygame.K_KP_ENTER:
            b_speed_y=y
            b_speed_x=x
            
        if event.type == pygame.KEYDOWN and event.key == pygame.K_LEFT:
            pad_rec.x += - 40
            if pad_rec.x < 0:
                pad_rec.x = 0
        
        if event.type == pygame.KEYDOWN and event.key == pygame.K_RIGHT:
            pad_rec.x += 40
            if pad_rec.x > width - pad_width:
                pad_rec.x = width - pad_width
            
    return(b_speed_x, b_speed_y)

def exitwind(ball_rec, width, b_speed_x, b_speed_y):
    if ball_rec.right > width:
        b_speed_x *= -1
    if ball_rec.left <0:
        b_speed_x *= -1
    if ball_rec.top <0:
        b_speed_y *= -1
    return(b_speed_x, b_speed_y)

def tail(wind, pad, pad_rec, bal, ball_rec):
    wind_color=(234, 214, 176)
    wind.fill(wind_color)
    wind.blit(pad, pad_rec)
    wind.blit(bal, ball_rec)
    #wind.blit(leg, leg_rec)
    pygame.display.update()
    
def message(font, msg, color, pos):
    text=font.render(msg, True, color)
    wind.blit(text, pos)
    
```
