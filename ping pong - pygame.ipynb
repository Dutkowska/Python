```python
import pygame, sys, os
from pygame.locals import *

def windows():
    size=(1000, 500)
    width=size[0]
    height=size[1]
    wind=pygame.display.set_mode((width, height))
    pygame.display.set_caption("Ping Pong")
    return(wind, width, height)
    
def padel():
    pad_width=100
    pad_height=10
    pad=pygame.Surface([pad_width, pad_height])
    pad_col=(0,98,98)
    pad.fill(pad_col)
    
    pad_rec=pad.get_rect()
    pad_pos=(450, 450)
    pad_rec.x=pad_pos[0]
    pad_rec.y=pad_pos[1]
    return(pad, pad_width, pad_rec, pad_rec.x)

def exitwind(ball_rec, width, b_speed_x, b_speed_y):
    if ball_rec.right > width:
        b_speed_x *= -1
    if ball_rec.left <0:
        b_speed_x *= -1
    if ball_rec.top <0:
        b_speed_y *= -1
    return(b_speed_x, b_speed_y)

def tail(wind, pad, pad_rec, bal, ball_rec):
    wind_color=(234, 214, 176)
    wind.fill(wind_color)
    wind.blit(pad, pad_rec)
    wind.blit(bal, ball_rec)
    pygame.display.update()
```

```python
import pygame, sys, os
from pygame.locals import *

windows()
wind, width, height=windows()
padel()
pad, pad_width, pad_rec, pad_rec.x=padel()

b_width=15
b_height=15
b_speed=(5,5)
b_speed_x=b_speed[0]
b_speed_y=b_speed[1]
b_color=(0,0,0)
bal=pygame.Surface([b_width, b_height], pygame.SRCALPHA).convert_alpha()
pygame.draw.ellipse(bal, b_color, [0,0,b_width, b_height])

ball_rec=bal.get_rect()
ball_rec.x=width/2
ball_rec.y=0

#ustawienia animacji
FPS=30 #liczba klatek na sek
fpsClock=pygame.time.Clock() #zegar śledzący czas

while True:
    for event in pygame.event.get():
        if event.type==QUIT:
            pygame.quit()
            sys.exit()
            
        if event.type==MOUSEMOTION:
            mx, my=event.pos
            pos=mx-(pad_width /2)

            if pos>width - pad_width: #right
                pos=width-pad_width
            if pos<0: #left
                pos=0

            pad_rec.x=pos
            
    #ball move
    ball_rec.move_ip(b_speed_x, b_speed_y)
    
    #exit from the window - left/right/top
    b_speed_x, b_speed_y = exitwind(ball_rec, width, b_speed_x, b_speed_y)
        
        #bottom - start new game
    if ball_rec.bottom >= height:
        ball_rec.x = width/2
        ball_rec.y = 0
    
    if ball_rec.colliderect(pad_rec):
        b_speed_y *= -1
    
    tail(wind, pad, pad_rec, bal, ball_rec)
    fpsClock.tick(FPS)
    
            
            
```
