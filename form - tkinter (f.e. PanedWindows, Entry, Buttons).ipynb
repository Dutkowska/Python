```python
from tkinter import *

okn=PanedWindow(orient=VERTICAL)
okn.pack(fill=BOTH, expand=1)
# okn - window, gi - vertical panel, 
# li/ri - horizontal parts of vertical panels (left/right)

hello="Good morning.\nYou are in the simple form. \nThis is only a small attempt of using tkinter:\n(f.e. PanedWindows) \n___________________________________________"
g1=Label(okn, text=hello, justify=LEFT)
okn.add(g1)

g2=PanedWindow(orient=HORIZONTAL)
okn.add(g2)

left=PanedWindow(orient=VERTICAL)
g2.add(left)

l1=Label(left, text="Your sex: ", justify=RIGHT)
left.add(l1)

l2=Label(left, text="Write your name: ", justify=RIGHT)
left.add(l2)



right=PanedWindow(orient=VERTICAL)
g2.add(right)

r1=Menubutton(right, text="Select", relief=RAISED)
r1.grid()
r1.menu=Menu(r1)
r1["menu"]=r1.menu

fem=IntVar()
mal=IntVar()

r1.menu.add_checkbutton(label="female", variable=fem)
r1.menu.add_checkbutton(label="male", variable=mal)
r1.pack()
right.add(r1)

name=StringVar()
r2=Entry(right, bd=5, textvariable=name)
right.add(r2)


def ok():
    if  mal.get():
        gm="Good morning Mr " + str(name.get())
        g4.config(text=gm)
    if fem.get():
        gm="Good morning Mrs " + str(name.get())
        g4.config(text=gm)
    """else:
        gm="Select your sex (only one)"
        g4.config(text=gm)"""

g3=Button(okn, text="OK", command=ok)
okn.add(g3)

g4=Label(okn)
okn.add(g4)

g5=PanedWindow(orient=HORIZONTAL)
okn.add(g5)

left2=PanedWindow(orient=VERTICAL)
g5.add(left2)

l3=Label(left2, text="Write your age: ", justify=RIGHT)
left2.add(l3)

right2=PanedWindow(orient=VERTICAL)
g5.add(right2)

r3=Entry(right2, bd=5)
right2.add(r3)

mainloop()
```
